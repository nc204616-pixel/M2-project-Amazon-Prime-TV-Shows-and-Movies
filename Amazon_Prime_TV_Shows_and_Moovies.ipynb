## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Remove Duplicate Rows
print("Duplicates before cleaning:")
print("Titles:", titles.duplicated().sum())
print("Credits:", credits.duplicated().sum())

titles.drop_duplicates(inplace=True)
credits.drop_duplicates(inplace=True)

print("\nDuplicates after cleaning:")
print("Titles:", titles.duplicated().sum())
print("Credits:", credits.duplicated().sum())

# 2. Handling Missing Values

# Fill missing text fields with "Unknown"
text_cols = ['description', 'age_certification', 'genres', 'production_countries']
for col in text_cols:
    titles[col] = titles[col].fillna("Unknown")

# Fill numeric missing values with median
num_cols = ['runtime', 'imdb_score', 'imdb_votes', 'tmdb_popularity', 'tmdb_score', 'seasons']
for col in num_cols:
    titles[col] = titles[col].fillna(titles[col].median())

# credits dataset missing character names
credits['character'] = credits['character'].fillna("Unknown")


# 3. Fix Data Types

titles['release_year'] = titles['release_year'].astype(int)
titles['seasons'] = titles['seasons'].astype(int)
titles['runtime'] = titles['runtime'].astype(int)

# Role conversion to category
credits['role'] = credits['role'].astype("category")


# 4. Convert list-like string columns

import ast

list_cols = ['genres', 'production_countries']

for col in list_cols:
    titles[col] = titles[col].apply(
        lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith('[') else []
    )


# 5. Standardize text columns

titles['title'] = titles['title'].str.strip().str.title()
titles['type'] = titles['type'].str.upper()
credits['name'] = credits['name'].str.strip().str.title()


# 6. Merge Dataset (Optional for Analysis)

merged = titles.merge(credits, on='id', how='left')

# 7. Final Checks

print("\nFinal Titles Shape:", titles.shape)
print("Final Credits Shape:", credits.shape)
print("Merged Shape:", merged.shape)

print("\nMissing Values After Cleaning:")
print(titles.isnull().sum())
````

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code
#Pair Plot for numerical_variable
import seaborn as sns
import matplotlib.pyplot as plt

# Select numerical variables
numerical_variables = titles.select_dtypes(include=[np.number]).columns

# Pair plot with hue by content type, corner=True avoids duplicate plots
sns.pairplot(
    data=titles,
    vars=numerical_variables,
    hue='type',      # separates Movies vs Shows
    corner=True,     # shows only lower triangle
    diag_kind='kde', # use KDE for diagonal plots
    plot_kws={'alpha':0.6, 's':40}  # adjust transparency and marker size
)

plt.suptitle('Pair Plot of Numerical Variables by Content Type', fontsize=16, weight='bold', y=1.02)
plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

Using hue='type' separates Movies and TV Shows, making it easy to see differences between content types.
